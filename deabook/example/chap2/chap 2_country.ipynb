```python
# import packages
from deabook import DEA
from deabook.constant import CET_ADDI, RTS_VRS1, RTS_CRS, \
                        OPT_LOCAL

import pandas as pd
# import all data (including the contextual varibale)

kind = "country" # "province" or "country"

if kind == "province":
    dmuname = '省份'
    data = pd.read_excel(r"../../data\china data.xlsx").query('year>2015').reset_index(drop=True)
    region = data[["省份","group"]]
else:
    dmuname = '国家'
    data = pd.read_excel(r"../../data\oecd data.xlsx").query('year>2015').reset_index(drop=True)


yearid = data[['year',dmuname]]
yearid
```

## DEA

```python
# define and solve the test_DEAt model
def DEA_CRS():

    model = DEA.DEA2(data,sent = "K L E =Y",   gy=[0], gx=[1,0,0], rts=RTS_CRS,  baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res
DEA_CRS = DEA_CRS()
DEA_CRS
```

```python
def DEA_VRS():

    model = DEA.DEA2(data,sent = "K L E =Y",   gy=[0], gx=[1,0,0], rts=RTS_VRS1,  baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

DEA_VRS = DEA_VRS()
DEA_VRS
```

```python
data
```

```python
DEA_CRS = DEA_CRS.drop(columns = ['optimization_status','rho'])

DEA_VRS = DEA_VRS.drop(columns = ['optimization_status','rho'])


DEA_CRS =  DEA_CRS.rename(columns={'te':'PEO_CRS'})
DEA_VRS =  DEA_VRS.rename(columns={'te':'PEO_VRS'})


DEA_CRS2 = pd.concat([DEA_CRS,yearid],axis=1)
DEA_VRS2 = pd.concat([DEA_VRS,yearid],axis=1)
```

```python
DEA_CRS2.query('year==2017').sort_values('PEO_CRS')
```

```python
DEA_VRS2.query('year==2017').sort_values('PEO_VRS')
```

```python
DEA_CRS3 = DEA_CRS2.pivot_table(
    values=['PEO_CRS',],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```

```python
DEA_VRS3 = DEA_VRS2.pivot_table(
    values=['PEO_VRS',],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```

## DDF

```python
from deabook import DEA
from deabook.constant import CET_ADDI, RTS_VRS1, RTS_CRS, \
                        OPT_LOCAL
```

```python
def DDF_CRS():

    model = DEA.DDF2(data,sent = "K L E=Y",  gy=[0], gx=[1,0,0],rts=RTS_CRS, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

DDF_CRS=DDF_CRS()
DDF_CRS
```

### DDF的结果和DEA的结果是一样的，这说明两者都对

```python
def DDF_VRS():

    model = DEA.DDF2(data,sent = "K L E=Y",  gy=[0], gx=[1,0,0],rts=RTS_VRS1, baseindex=None,refindex=None)
    res = model.optimize(solver="mosek")
    # print(model.info(1))
    return res

DDF_VRS=DDF_VRS()
DDF_VRS
```

```python
DDF_CRS['EI_CRS'] =  DDF_CRS['tei']
DDF_VRS['EI_VRS'] = DDF_VRS['tei']

DDF_CRS = DDF_CRS.drop(columns = ['optimization_status','rho'])
DDF_VRS = DDF_VRS.drop(columns = ['optimization_status','rho'])



DDF_CRS2 = pd.concat([DDF_CRS,yearid],axis=1)
DDF_VRS2 = pd.concat([DDF_VRS,yearid],axis=1)
```

```python
DDF_CRS2
```

```python
DDF_CRS3 = DDF_CRS2.pivot_table(
    values=['EI_CRS',],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```

```python
DDF_VRS3 = DDF_VRS2.pivot_table(
    values=['EI_VRS',],
    index=dmuname,
    columns='year',
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
    observed=False,
)  
```
