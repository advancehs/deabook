## MQDEA

```python

# define and solve the test_DEAt model
def MQDEA_CRS():

    # model = DEA.DEA(data,sent = "K L E=Y:CO2",  orient="L", rts=RTS_VRS, baseindex=None,refindex=None)
    model = MQDEAweak.MQDEAweak(data,id=dmuname,year='year',sent = "K L E=Y:CO2",  gy=[0], gx=[0,0,0], gb=[1], rts=RTS_CRS,\
                        tech = CONTEMPORARY,solver="mosek")
    res = model.optimize()
    # print(model.info(1))
    return res
MQDEA_CRS = MQDEA_CRS()
MQDEA_CRS
```

```python

# define and solve the test_DEAt model
def MQDEA_VRS_SAME():

    # model = DEA.DEA(data,sent = "K L E=Y:CO2",  orient="L", rts=RTS_VRS, baseindex=None,refindex=None)
    model = MQDEAweak.MQDEAweak(data,id=dmuname,year='year',sent = "K L E=Y:CO2",  gy=[0], gx=[0,0,0], gb=[1], rts=RTS_VRS1,\
                        tech = CONTEMPORARY,solver="mosek")
    res = model.optimize()
    # print(model.info(1))
    return res
MQDEA_VRS_SAME = MQDEA_VRS_SAME()
MQDEA_VRS_SAME
```

```python

# define and solve the test_DEAt model
def MQDEA_VRS_DIFF():

    # model = DEA.DEA(data,sent = "K L E=Y:CO2",  orient="L", rts=RTS_VRS, baseindex=None,refindex=None)
    model = MQDEAweak.MQDEAweak(data,id=dmuname,year='year',sent = "K L E=Y:CO2",  gy=[0], gx=[0,0,0], gb=[1], rts=RTS_VRS2,\
                        tech = CONTEMPORARY,solver="mosek")
    res = model.optimize()
    return res
MQDEA_VRS_DIFF = MQDEA_VRS_DIFF()
MQDEA_VRS_DIFF
```
