```python
import numpy as np
from scipy.optimize import minimize

def u_minus(x, r1):
    return np.where(x < r1, x, 0)

def u_plus(x, r2):
    return np.where(x > r2, x, 0)

def W(X, Z, r1, r2):
    return np.column_stack((u_minus(X, r1), u_plus(X, r2), Z))

def objective(params, Y, W):
    beta1, beta2, theta = params[:2], params[2:-1], params[-1]
    eta = np.concatenate((beta1, beta2, [theta]))
    Y_bar = np.mean(Y)
    W_bar = np.mean(W, axis=0)
    return (Y - X @ eta - np.mean(Y - X @ eta)) @ (W - W_bar) - theta * (beta1 - beta2) @ (beta1 - beta2)

def estimate_params(Y, X, Z):
    bounds = [(None, None), (None, None)] + [(None, None)] * (Z.shape[1])
    result = minimize(objective, x0=[0, 0] + [0] * Z.shape[1], args=(Y, W(X, Z, 0, 0)), bounds=bounds)
    r1, r2 = minimize(lambda r: objective([*result.x[:2], *result.x[2:], r], Y, W(X, Z, r[0], r[1])) , x0=[np.min(X), np.max(X)], bounds=[(np.min(X), np.max(X)), (np.min(X), np.max(X))]).x
    eta = np.linalg.inv(W(X, Z, r1, r2).T @ W(X, Z, r1, r2)) @ W(X, Z, r1, r2).T @ Y
    return eta, r1, r2

X = np.linspace(-5, 5, 100)
Z = np.random.normal(size=(100, 3))
Y = 2 * u_minus(X, -1) + 3 * u_plus(X, 1) + Z @ np.array([1, 2, 3]) + np.random.normal(size=100)
eta, r1, r2 = estimate_params(Y, X, Z)
print(f"Estimated parameters: {eta}, r1: {r1}, r2: {r2}")

```
